#  Model Building:

## 1)Support Vector Machine Algorithm:
Support Vector Machine or SVM is one of the most popular Supervised Learning algorithms, which is used for Classification problems.
The goal of the SVM algorithm is to create the best line or decision boundary that can segregate n-dimensional space into classes so that we can easily put the new data point in the correct category in the future. This best decision boundary is called a hyperplane.
SVM chooses the extreme points/vectors that help in creating the hyperplane. These extreme cases are called as support vectors, and hence algorithm is termed as Support Vector Machine.
*********************************
To apply SVM to non-linear problem we have to use a kernel.
## what is a kernerl and how it works?
A kernel is a function that takes the original non-linear problem and transforms it into a linear one within the higher-dimensional space.  
## How to transform this problem into a linear one?
Radial Basis Function (RBF) kernel is the default kernel used within the sklearn’s SVM classification algorithm.
